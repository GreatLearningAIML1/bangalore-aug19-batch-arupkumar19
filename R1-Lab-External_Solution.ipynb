### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
Sepal_Length_Range = max(df['Sepal Length (in cm)']) - min(df['Sepal Length (in cm)'])
print('Range of Sepal Length (in cm) column :', Sepal_Length_Range)

Sepal_Width_Range = max(df['Sepal Width in (cm)']) - min(df['Sepal Width in (cm)'])
print('Range of Sepal Length (in cm) column :', Sepal_Width_Range)

Petal_length_Range = max(df['Petal length (in cm)']) - min(df['Petal length (in cm)'])
print('Range of Sepal Length (in cm) column :', Petal_length_Range)

Petal_width_Range = max(df['Petal width (in cm)']) - min(df['Petal width (in cm)'])
print('Range of Sepal Length (in cm) column :', Petal_width_Range)

SepalExtra_Range = max(df['SepalExtra']) - min(df['SepalExtra'])
print('Range of Sepal Length (in cm) column :', SepalExtra_Range)
```
